```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../data/clean/all/clean_pokemon_data_all.csv')
```

```python
df.head()
```

# Visualization

```python
sns.regplot(data=df[df['type_1'] == 'grass'], x='height', y='weight',)
```

```python
sns.regplot(data=df[df['type_1'] == 'fire'], x='height', y='weight',)
```

```python
sns.swarmplot(data=df, x='type_1', y='hp', palette='Set2')
```

```python
sns.regplot(data=df, x='hp', y='weight')
```

```python
sns.regplot(data=df, x='defense', y='weight')
```

```python
sns.regplot(data=df, x='special-defense', y='defense')
```

```python
sns.scatterplot(data=df, x='height', y='weight', hue='type_1', style='type_1', s=100)
plt.title('Relation entre la taille et le poids des Pokmon')
plt.xlabel('Taille (m)')
plt.ylabel('Poids (kg)')
plt.legend(title='Type')
plt.show()
```

```python
sns.lmplot(data=df, x='height', y='weight', hue='type_1', palette='Set2')
plt.title('Rgression linaire entre la taille et le poids des Pokmon')
plt.xlabel('Taille (m)')
plt.ylabel('Poids (kg)')
plt.show()
```

```python
sns.boxplot(data=df, x='type_1', y='weight', palette='Set2')
plt.title('Distribution du poids par type de Pokmon')
plt.xlabel('Type de Pokmon')
plt.ylabel('Poids (kg)')
plt.show()
```

```python
sns.violinplot(data=df, x='type_1', y='height', palette='Set2')
plt.title('Distribution de la taille par type de Pokmon')
plt.xlabel('Type de Pokmon')
plt.ylabel('Taille (m)')
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de corrlation des statistiques des Pokmon')
plt.show()
```
